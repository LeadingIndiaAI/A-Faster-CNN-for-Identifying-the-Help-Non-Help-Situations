## Project Understanding

### What exactly is the problem?

Binary image classification problem. Identify the presence of help or not help situation from the images scrapped from web having various resolutions. The key challenges for classification problem are as follows:
1. All the images should scaled down to same size
2. There are water marks in the image that might affect the classification accuracy of the model
3. There is lot of variety in terms of features that the model should be able to learn

## Data Understanding

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import pandas as pd
import os
from PIL import Image
from tqdm import tqdm
import seaborn as sns
import statistics
import pathlib
```

```python
# Distribution of data
help_data_path = 'DataSet-SAR/help'
nothelp_data_path = 'DataSet-SAR/nothelp'

count_help_data = os.listdir(help_data_path)
count_not_help_data = os.listdir(nothelp_data_path)

print(f'Number of data points for "help" class is : {len(count_help_data)}')
print(f'Number of data points for "Not help" class is : {len(count_not_help_data)}')

x = ['help', 'not_help']
height = [len(count_help_data), len(count_not_help_data)]
total_images = len(count_help_data) + len(count_not_help_data)
sizes = [(x/total_images) * 100 for x in height] 

figure(num=None, figsize=(7, 7), dpi=100, facecolor='w', edgecolor='k')
plt.subplot(1,2,1)
plt.bar(x, height, width=0.5, color = 'orange')
plt.xticks(x)
plt.title("Data Distribution Plot")
plt.ylabel("Count")
plt.subplot(1,2,2)
explode = (0.1, 0)
plt.pie(sizes,labels=x, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Pie Chart")
# plt.ylabel("Count")
plt.axis('equal')
plt.subplots_adjust(wspace =0.5)
plt.show()
```

**As it is clear from the above charts that the data points for classes 'help' and 'not help' are equal hence there is less probablity of model being biased towards sigle class**

```python
# Image resolution plots
height_1 = [] # for help images
width_1 = [] 
height_2 = [] #For not help images
width_2 = []
for i in tqdm(range(len(count_help_data))):
    height_1.append(Image.open(help_data_path + '/'+str(count_help_data[i])).size[1])
    width_1.append(Image.open(help_data_path +'/'+str(count_help_data[i])).size[0])
    height_2.append(Image.open(nothelp_data_path+ '/'+str(count_not_help_data[i])).size[1])
    width_2.append(Image.open(nothelp_data_path +'/'+str(count_not_help_data[i])).size[0])

    
```

```python
figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')
plt.subplot(2,2,1)
plt.plot(height_1, "bo")
plt.plot(statistics.mean(height_1), "ro")
plt.title("Heights of images belonging to 'Help' class")
plt.ylabel("Count")
plt.subplot(2,2,2)
plt.plot(width_1, "r*")
plt.plot(statistics.mean(width_1), "b*")
plt.title("Widths of images belonging to 'Help' class")
plt.ylabel("Count")
plt.subplot(2,2,3)
plt.plot(height_2, "bo")
plt.plot(statistics.mean(height_2), "ro")
plt.title("Heights of images belonging to 'Not Help' class")
plt.ylabel("Count")
plt.subplot(2,2,4)
plt.plot(width_2, "r*")
plt.plot(statistics.mean(width_2), "b*")
plt.title("Widths of images belonging to 'Not Help' class")
plt.ylabel("Count")
plt.show()

print(f'Average height and width for "help" dataset is : {int(statistics.mean(height_1)),int(statistics.mean(width_1))}')
print(f'Average height and width for "Not help" dataset is : {int(statistics.mean(height_2)),int(statistics.mean(width_2))}')
```

```python
figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')
plt.subplot(2,2,1)
sns.distplot(height_1, color = 'salmon')
plt.xlabel("height")
plt.ylabel("frequency")
plt.title("Distribution of heights --'Help'")
plt.subplot(2,2,2)
sns.distplot(width_1, color='slateblue')
plt.xlabel("width")
plt.ylabel("frequency")
plt.title("Distribution of Widths --'Help'")
plt.subplot(2,2,3)
sns.distplot(height_2,  color = 'salmon')
plt.xlabel("height")
plt.ylabel("frequency")
plt.title("Distribution of heights --'Not Help'")
plt.subplot(2,2,4)
sns.distplot(width_2, color='slateblue')
plt.xlabel("width")
plt.ylabel("frequency")
plt.title("Distribution of Widths --'Not Help'")
plt.subplots_adjust(wspace =0.5, hspace=0.25)
plt.show()
```

### Chek for image compression format

```python
help_extns = []
not_help_extns = []
for i in tqdm(range(len(count_help_data))):
    help_extns.append(pathlib.Path(help_data_path + '/'+str(count_help_data[i])).suffix)
    not_help_extns.append(pathlib.Path(nothelp_data_path+ '/'+str(count_not_help_data[i])).suffix)
    
print(set(help_extns))
print(set(not_help_extns))   
```

**All the images in both the classes have same compression format - jpg**

## rename images

```python
index = 1
for dirname in os.listdir("DataSet-SAR/"):
    if os.path.isdir('DataSet-SAR/'+dirname):
        for filename in tqdm(os.listdir('DataSet-SAR/'+dirname)):
            if filename.endswith('.jpg'):
                os.rename('DataSet-SAR/'+dirname + "/" + filename, 'DataSet-SAR/'+dirname + "/" + 'image_'+str(index) + ".jpg")
                index += 1
```

```python
from natsort import natsorted
img_names = os.listdir(help_data_path)
labels = np.concatenate((np.ones((2078)),np.zeros((2078))))
img_names[len(img_names):] = os.listdir(nothelp_data_path)
img_names = natsorted(img_names)

# prepare dataframe with image name and classlabels
dataframe = pd.DataFrame()
dataframe['id'] = img_names
dataframe['class'] = labels
print(dataframe.head())

#Save dataframe to csv
dataframe.to_csv('datset.csv', index=False)
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_temp, y_train, y_temp = train_test_split(img_names,labels,stratify = labels,test_size=0.2, random_state=42)


X_test, X_validation, y_test, y_validation  = train_test_split(X_temp,
                                                                             y_temp,
                                                                             stratify = y_temp,
                                                                             test_size=0.35, 
                                                                             random_state=42)
print(f'Number of images in X_train_final is : {len(X_train)}')
print(f'Number of images in y_train_final is : {len(y_train)}')
print(f'Number of images in X_validation is : {len(X_validation)}')
print(f'Number of images in y_validation is : {len(y_validation)}')
print(f'Number of images in X_test  is : {len(X_test)}')
print(f'Number of images in y_test  is : {len(y_test)}')


figure(num=None, figsize=(10, 7), dpi=100, facecolor='w', edgecolor='k')
plt.subplot(2,3, 1)
plt.bar(x,[sum(y_train), len(y_train) - sum(y_train)], color = 'orange')
plt.title("Distribution in Training dataset")
plt.ylabel("Count")

plt.subplot(2,3, 2)
plt.bar(x,[sum(y_validation), len(y_validation) - sum(y_validation)], color = 'gold')
plt.title("Distribution in Validation dataset")
plt.ylabel("Count")

plt.subplot(2,3, 3)
plt.bar(x,[sum(y_test), len(y_test) - sum(y_test)], color = 'steelblue')
plt.title("Distribution in Test dataset")
plt.ylabel("Count")
plt.subplots_adjust(wspace=2.5)
plt.show()
```

```python
figure(num=None, figsize=(5, 5), dpi=100, facecolor='w', edgecolor='k')
total_data_size = len(y_train) + len(y_validation) + len(y_test)
data_size = [(len(y_train)/total_data_size) * 100 , (len(y_validation)/total_data_size) * 100, (len(y_test)/total_data_size) * 100]
explode = (0, 0, 0)
plt.pie(data_size,labels=['Training data','validation data','Test data'], explode=explode, autopct='%1.1f%%', shadow=True, startangle=90,
       colors= ['mediumaquamarine', 'gold', 'lightsalmon'])
plt.title("Data Split")
plt.axis('equal')
plt.show()
```

```python
import csv
with open('train.csv', 'w', newline='') as myfile:
    wr = csv.writer(myfile)
    for i in range(len(y_train)):
        temp = [X_train[i], y_train[i]]
        wr.writerow(temp)
        
with open('val.csv', 'w', newline='') as myfile:
    wr = csv.writer(myfile)
    for i in range(len(y_validation)):
        temp = [X_validation[i], y_validation[i]]
        wr.writerow(temp)
        
        
with open('test.csv', 'w', newline='') as myfile:
    wr = csv.writer(myfile)
    for i in range(len(y_test)):
        temp = [X_test[i], y_test[i]]
        wr.writerow(temp)
```

```python
def change(x):
    if x==0:
        return 'nothelp'
    return 'help'
train['labels'] = train['labels'].apply(change)
val['labels'] = val['labels'].apply(change)
test['labels'] = test['labels'].apply(change)
```

```python
from ipywidgets import IntProgress
from IPython.display import display
import time
```

```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
def move_img(df, loc):
    max_count = len(df)
    f = IntProgress(min=0, max=max_count)
    display(f)
    for ind in df.index:
        img=df['img'][ind]
        label=df['labels'][ind]
        src = f'DataSet-SAR/{label}/{img}'
        image = Image.open(src).convert('RGB')
        image.save(f'data/{loc}/{label}/{img}')
        f.value+=1
```

```python
move_img(train, 'train')
move_img(val, 'val')
move_img(test, 'test')
```
